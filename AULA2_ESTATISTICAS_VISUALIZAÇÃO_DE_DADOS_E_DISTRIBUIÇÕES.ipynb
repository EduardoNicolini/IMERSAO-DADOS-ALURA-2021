```python
#bibliotecas 
import pandas as pd

#URL coma base de dados
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

# Faz a leitura do arquivo zip
dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

```python
# Modificando o nome da coluna da base de dados 
mapa = {'droga': 'composto'}# string onde vol passar a coluna a ser alterada e o nome a ser alterado 
dados.rename(columns=mapa, inplace=True)
```

```python
# Realiso a leitura da base de dados
dados.head()
```

```python
# Na frente do value_counts() Add o [0:5] para selecionar o numero de colunas a ser exibido [começa daqui : vai até aqui]
compostos = dados['composto'].value_counts()[0:5] # Armazeno esse valor dentro da variavel compostos
compostos# imprimo o valor armazenado na variavel
```

```python
compostos.index # Me traz um array com os valores selecionados
```

```python
# Posso tambem selecionar as colunas desejadas e trazer os valores dentro de um array
cod_compostos = dados['composto'].value_counts().index[0:5]

cod_compostos # imprimo o valor armazenado na variavel
```

```python
# Retorna tudo o que tiver dentro de composto e tudo que esteja igual ou 'in' a cod_composto
dados.query('composto in @cod_compostos')
```

Grafico usando seaborn
 https://seaborn.pydata.org/index.html

```python
#Biblioteca seaborn
import seaborn as sns

# Biblioteca para plotar o grafico
from matplotlib import pyplot as plt

sns.set()# Salva as configurações padrões de imagens do seaborn

plt.figure(figsize= (8, 6))# Defino o tamanho do grafico
plt.rcParams['xtick.labelsize'] = 13 # altera o tamanho da fonte da coluna x
plt.rcParams['ytick.labelsize'] = 13 # altera o tamanho da fonte da coluna y

# Usando o @ eu aviso a string dentro da query que esse cod_composto ja foi declarado pelo phyton fora da string 
ax= sns.countplot( x = 'composto', data = dados.query('composto in @cod_compostos')) # Armazeno o grafico dentro da variavel padrão 'ax'

ax.set_title('Top 5 Compostos', fontsize = 18) # Aqui chamo a variavel 'ax' e passo o parametro para inserir o titulo

ax.set(ylabel="contagem") # Edita a legenda no eixo y, Para alterar o valor no eixo x , xlabel= "digite a legenda aqui"

plt.show() # limpa informações desnecessarias do grafico
```

* Média destas amostras;
* Min: Menor valor da série;

**Desafio 06: Fazer outras análises com o boxplot e com o histograma**

```python
dados.loc[:, 'c-0':'c-99'].describe().T['mean'].hist(bins=50)
plt.show()
```

```python
#Histograma para visualizar o valor médio do 3º quarto de cada coluna entre c-0 até c-99 no seaborn

columns_third_quarter = dados.loc[:, 'c-0':'c-99'].describe().T['75%']
sns.set
ax = sns.histplot(data = columns_third_quarter)
```

```python
#Boxplot para visualizar a relação entre os valores da coluna g-0 no eixo y e da coluna dose no eixo x, onde a media das doses aplicadas D1 e D2 se encontra no valor 0.
sns.boxplot(y = 'g-0', x = 'tratamento', data = dados)
plt.show()
```

```python
#Boxplot para visualizar a relação entre os valores da coluna g-0 no eixo y e da coluna dose no eixo x, onde a media das doses aplicadas D1 e D2 se encontra no valor 0.
sns.boxplot(y = 'g-0', x = 'dose', data = dados)
plt.show()
```
