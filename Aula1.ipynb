```python
#bibliotecas 
import pandas as pd

# Biblioteca para plotar o grafico
from matplotlib import pyplot as plt



```

```python
#URL coma base de dados
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

# Faz a leitura do arquivo zip
dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

```python
# Enumera Linhas e colunas
dados.shape
```

```python
# Pega um unico valor na coluna 'droga'
dados['droga'].unique()
```

```python
# Normaliza e conta os valores que estão dentro da coluna
dados['tratamento'].value_counts(normalize = True)
```

**Desafio 05: Procurar na documentação o método query(pandas)**

Você pode consultar as colunas de um DataFrame com uma expressão booleana.

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.query.html

```python
# Consulta somente o tratamento com drogas somente com a dose D2 
dados.query('tratamento == "com_droga" & dose == "D2" ')
```

**Desafio 06: Renomear as colunas tirando o hífen**

```python
# Filtra as colunas e retira os - 
dados.columns = dados.columns.str.replace('-','')
# Le o arquivo sem o hífen
dados.head()
```

**Desafio 07: Deixar os gráficos bonitões. (Matplotlib.pyplot)**

```python
dados['tempo'].value_counts().plot.bar(color=['darkblue','gold','cyan'])

plt.title('Gráfico Bonitão (Matplotlib.pyplot)')
plt.xlabel("Horas Submetidas")
plt.ylabel("Quantidades")
```

```python
eixoX = ['Com droga', 'Com comtrole']
eixoY = dados.tratamento.value_counts()

plt.xlabel('Grupos de tratamento', fontsize = 14)
plt.ylabel('Números de experimentos', fontsize = 14)
plt.title('Tipos de tratamento', fontsize = 14)

plt.bar(eixoX, eixoY, color=['darkblue','gold','cyan'])
plt.show()
```
